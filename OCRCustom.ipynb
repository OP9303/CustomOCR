```python
img = cv2.imread('/content/1.png')
```

```python
cv2_imshow(img) #BlueGreenRed Format for images opened in pytesseract
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img)
```

```python
text = pytesseract.image_to_string(rgb) #extraction of text from image using pytesseract
print(text) #printing of text extracted from image
```
